## While loops

```python
from datetime import datetime
```

```python
datetime.now().second
```

```python
wait_unitl = datetime.now().second + 2

while datetime.now().second != wait_unitl:
    print('Still Waiting!')

print(f'we are at {wait_unitl} seconds!')
```

## Pass

```python
wait_unitl = datetime.now().second + 2

while datetime.now().second != wait_unitl:
    pass

print(f'we are at {wait_unitl} seconds!')
```

## Break

```python
wait_unitl = datetime.now().second + 2

while True:
    if datetime.now().second == wait_unitl:
        print(f'we are at {wait_unitl} seconds!')
        break
    
```

## Continue

```python
wait_unitl = datetime.now().second + 2

while datetime.now().second != wait_unitl:
    continue
    print('Still Waiting!')
    

print(f'we are at {wait_unitl} seconds!')
```

```python
wait_unitl = datetime.now().second + 2

while True:
    if datetime.now().second < wait_unitl:
        continue
    break
    

print(f'we are at {wait_unitl} seconds!')
```
