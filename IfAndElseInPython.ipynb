### If statements with "FizzBuzz"

```python
for n in range(1,101):
    if n % 15 == 0:
        print('FizzBuzz')
    else:
        if n % 3 == 0:
            print('Fizz')
        else:
            if n % 5 == 0:
                print('Buzz')
            else:
                print(n)
            
```

### Elif statements

```python
for n in range(1,101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    else:
        print(n)
```

### Single Line If Statements

```python
n = 5

print('Fizz' if n % 3 == 0 else n)
```
